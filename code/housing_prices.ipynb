```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/State_Zhvi_3bedroom.csv', index_col='RegionName')
```

```python
df.dropna(inplace=True)
df.columns
months = df.columns[2:]
```

```python
#df.drop(['RegionID', 'SizeRank'], axis=1, inplace=True)
```

```python
monthly_avg = []
for month in months:
    #print(month, df[month].mean())
    monthly_avg.append(df[month].mean())
    print(month, df[month].mean(), df[month].min(), df[month].max(), df[month].std())
```

```python
plt.plot(months, monthly_avg)
```

```python
plt.plot(months, df.loc['Michigan'][2:])
```

```python
time_labels = pd.date_range('1996-04-01', '2018-10-01', freq='12M')
```

```python
time_labels
```

```python
fig, ax = plt.subplots()
ax.plot_date(months, df.loc['Michigan'][2:])
ax.xaxis.set_major_locator(time_labels)
fig.autofmt_xdate()
plt.show()
```
